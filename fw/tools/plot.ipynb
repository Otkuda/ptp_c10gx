```python
import csv
import matplotlib.pyplot as plt

```

```python
# Укажите путь к вашему файлу
filename = 'res.csv'

# Список для хранения значений первого столбца
values = []
avg_values = []

# Чтение CSV-файла
with open(filename, newline='') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row:  # Проверяем, что строка не пустая
            try:
                values.append(float(row[0]))
            except ValueError:
                pass  # Пропустить заголовки или некорректные значения

for i in range(len(values)-11):
    avg_values.append(sum(values[i:i+10])/10)

# Построение графика
plt.figure(figsize=(10, 6))
plt.ylim(0.00095, 0.00099)
plt.xlabel('t, сек')
plt.ylabel('$\Delta$, сек')
plt.title('Задержка между PPS')
plt.grid(True)


plt.plot(values, label='Расстояние PPS')
plt.plot(avg_values, 'r', label='Скользящее среднее')
plt.legend()
plt.show()
```
